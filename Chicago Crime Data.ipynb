```python


chicago_full = pd.read_csv(RAW_FILE)
chicago_full
```

```python
# this cell can take up to 1 min to run
date_format = "%m/%d/%Y %H:%M:%S %p"

chicago_full['Datetime'] = pd.to_datetime(chicago_full['Date'], format=date_format)
chicago_full = chicago_full.sort_values('Datetime')
chicago_full
```

### Separate the Full Dataset by Years

```python
chicago_full['Datetime'].dt.year
```

```python
# save the years for every crime
chicago_full["Year"] = chicago_full['Datetime'].dt.year.astype(str)
chicago_full["Year"].value_counts()
```

```python
## Dropping unneeded columns to reduce file size
drop_cols = ["X Coordinate","Y Coordinate", "Community Area","FBI Code",
             "Case Number","Updated On",'Block','Location','IUCR']
```

```python
# save final df
chicago_final = chicago_full.drop(columns=drop_cols)
chicago_final = chicago_final.set_index('Datetime')
chicago_final
```

```python
# unique # of year bins
year_bins = chicago_final['Year'].astype(str).unique()
year_bins
```

```python
FINAL_DROP = ['Year']
```

```python
## set save location 

os.makedirs(OUTPUT_FOLDER, exist_ok=True)
print(f"[i] Saving .csv's to {OUTPUT_FOLDER}")
## loop through years
for year in year_bins:
    
    ## save temp slices of dfs to save.
    temp_df = chicago_final.loc[year]
    temp_df = temp_df.sort_index()
    temp_df = temp_df.reset_index(drop=True)
    temp_df = temp_df.drop(columns=FINAL_DROP)

    # save as csv to output folder
    fname_temp = f"{OUTPUT_FOLDER}Chicago-Crime_{year}.csv"#.gz
    temp_df.to_csv(fname_temp,index=False)

    print(f"- Succesfully saved {fname_temp}")
```

### Analysis 1:



- Comparing Police Districts:
    
    - Which district has the most crimes? Which has the least?

```python
district_crime= df['District'].value_counts(normalize=True)

district_crime.sort_values().plot(title='Crime Count by District', ylabel='District',
                                  kind='barh');
```

```python
district_counts= df['District'].value_counts().to_frame('Crime Count')
district_counts
district_counts.style.bar('Crime Count')
```

- While District 21 has the least amount of crime with only 4.

### Analysis 2:

- Crimes Across the Years:

    - Is the total number of crimes increasing or decreasing across the years?

    - Are there any individual crimes that are doing the opposite (e.g decreasing when overall crime is increasing or vice-versa)?

```python
df_yrs= df.resample('A').size()
df_yrs
```

```python
df_yrs.plot(title='Crime Number Total', ylabel='Crimes');
```

- Total number of crime as you can see in the graph above has decreased throughout the years. 

```python
crimes_by_year= df.groupby('Primary Type')['Year'].value_counts().sort_index()

crimes_by_year
```

```python
dfyearly= crimes_by_year.to_frame('Number of Crimes').reset_index()
```

```python
cby= sns.catplot(data=dfyearly, x='Year', y='Number of Crimes',
                kind='point', aspect= 3, height= 6,
                hue= 'Primary Type', col= 'Primary Type', col_wrap=2,
                sharey=False, sharex= False)
[ax.grid(axis='x') for ax in cby.axes];
```

- While overall number of crime has decreased throughout the years, some types of crime has increased in the same timespan. 


- **Crimes that have increased thru the years:**
    - Concealed Carry License Violation
    - Criminal Sexual Assault
    - Deceptive Practice
    - Homicide
    - Human Trafficking
    - Obscenity
    - Stalking
    - Weapons Violation

### Analysis 3:


-  Comparing Holidays:
    - Are there any holidays that show an increase in the # of crimes?
    - Are there any holidays that show a decrease in the # of crimes?

```python
hol_crimes= df.groupby('US Holiday')['Year'].value_counts().sort_index()
hol_crimes= hol_crimes.to_frame('Number of Crimes').reset_index()
hol_crimes
```

```python
hol= sns.catplot(data=hol_crimes, x='Year', y='Number of Crimes',
                kind='point', aspect= 3, height= 6,
                hue= 'US Holiday', col= 'US Holiday', col_wrap=2,
                sharey=False, sharex= False)
[ax.grid(axis='x') for ax in hol.axes];
```

- Holiday Observations:
    - All Holiday Crimes are decreasing by year except for **one** holiday. 
    - Juneteenth is the only holiday that shows an increase in crime by year. 
        - **NOTE:** We only have 3 Years of data for the holiday, Juneteenth.
